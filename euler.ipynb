```python
import math
import matplotlib.pyplot as plt

h = 0.1
x = 0
y = 1

xs = [x]
ys = [y]

for i in range(30):
    y = y + h * (-x + y)
    x = x + h
    xs.append(x)
    ys.append(y)

x_exact = []
y_exact = []
xx = 0
while xx <= xs[-1]:
    x_exact.append(xx)
    y_exact.append(-xx - 1 + 2*math.exp(xx))
    xx += 0.01

plt.plot(x_exact, y_exact)
plt.plot(xs, ys, "o--")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()

```
